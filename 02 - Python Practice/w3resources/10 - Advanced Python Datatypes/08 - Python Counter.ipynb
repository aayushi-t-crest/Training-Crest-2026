## 1. Counter of Vowels

Write a Python program that creates a counter of the vowels in the word "Python Exercises"

```python
from collections import Counter
```

```python
word = "Python Exercises"
vowels = "aeiouAEIOU"
```

```python
vowel_cnt= Counter(c for c in word if c in vowels)

print("Vowel Counts:")
for vowel,count in vowel_cnt.items():
    print(f"{vowel}:{count}")
```
