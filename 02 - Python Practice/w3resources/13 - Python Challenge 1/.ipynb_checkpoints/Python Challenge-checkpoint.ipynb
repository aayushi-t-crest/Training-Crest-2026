####  **Write a Python program to check if a given positive integer is a power of three**
Input : 9
Output : True

```python
def is_Power_of_three(n):
     while (n % 3 == 0):
         n /= 3;         
     return n == 1

print(is_Power_of_three(9))
print(is_Power_of_three(81))
print(is_Power_of_three(21))       
```

#### **Find the single number in a list that doesn't occur twice**

```python
def single_number(arr):
    result = 0
    for i in arr:
        result ^= i
    return result

arr1 = [5, 3, 4, 3, 4]
arr2 = [3, 2, 5, 2, 1, 5, 3]

print(single_number(arr1))
print(single_number(arr2))
```

#### **Write a Python program to find a missing number from a list**
Input : [1,2,3,4,6,7,8]
Output : 5

```python
def missing_number(num_list):
    return sum(range(num_list[0],num_list[-1]+1)) - sum(num_list)

print(missing_number([1,2,3,4,5,6,7,8,10]))

print(missing_number([10,11,12,14,15,16,17]))
```
