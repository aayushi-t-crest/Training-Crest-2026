#### **Write a Python program to search the country name from given state name using Nominatim API and GeoPy package**

```python
from geopy.geocoders import Nominatim

def get_country_from_state(state_name):
    
    geolocator = Nominatim(user_agent="state_to_country_app")

    # Geocode the state name
    location = geolocator.geocode(state_name, addressdetails=True)

    if location:
        address = location.raw.get("address", {})
        country = address.get("country", "Country not found")
        return country
    else:
        return "Location not found"


state = input("Enter state name: ")
country = get_country_from_state(state)

print("Country:", country)
```
