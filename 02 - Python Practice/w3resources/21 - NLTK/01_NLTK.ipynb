## Write a Python NLTK program to remove stop words from a given text.

```python
from nltk.corpus import stopwords
stoplist = stopwords.words('english')
```

```python
text = '''
This is a simple paragraph that is written to show how stopwords appear in normal text. 
The words like is, a, the, and, to, of, in, and for are used very frequently in sentences, 
but they do not carry important meaning. When we process text in natural language processing,
these stopwords are often removed because they do not add much value to the overall 
understanding of the sentence.
'''
```

```python
print("\nOriginal string:")
print(text)
```

```python
clean_word_list = [word for word in text.split() if word not in stoplist]
print("\nAfter removing stop words from the said text:")
print(clean_word_list)
```
