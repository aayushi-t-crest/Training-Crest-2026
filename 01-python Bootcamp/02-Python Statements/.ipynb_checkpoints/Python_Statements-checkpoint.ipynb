## While loops in python

````python
x = 0

while x < 5:
    print(f'The current of x is {x}')
    x = x+1
````

### break, continue, pass

we can use brreak, continnuw, and pass statements in our loops to add additinonal functionality

break: Breaks out of the current closest enclosing loop.<br>
continue: Goes to the top of the closest enclosing loop.<br>
pass: Does nothing at all.

````python
x = [1,2,3]

for item in x:

    pass


print('end of script')
````

````python
mystring = 'Sammy'

for letter in mystring:
    if letter == 'a':
        break;
    print(letter)
````

````python
x = 0

while x<5:
    if(x==2):
        break
    print(x)
    x = x+1
````

### enumerate

````python
index_count = 0

for letter in 'abcde':
    print("At index {} the letter is {}".format(index_count,letter))
    index_count += 1
````

````python
for i,letter in enumerate('abcde'):
    print("At index {} the letter is {}".format(i,letter))
````

# List comprehension

````python
lcomp = [x for x in 'word'] 
lcomp
````

````python
lst = [x**2 for x in range(0,11)]
lst
````

````python
lst = [x for x in range(11) if x % 2 == 0]
lst
````

````python
celsius = [0,10,20.1,34.5]

fahrenheit = [((9/5)*temp + 32) for temp in celsius ]

fahrenheit
````

````python
lst = [ x**2 for x in [x**2 for x in range(11)]]
lst
````
