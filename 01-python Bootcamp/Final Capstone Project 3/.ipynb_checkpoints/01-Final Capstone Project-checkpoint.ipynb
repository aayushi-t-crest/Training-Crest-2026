#### Factorial Finder - The Factorial of a positive integer, n, is defined as the product of the sequence n, n-1, n-2, ...1 and the factorial of zero, 0, is defined as being 1. Solve this using both loops and recursion.

### Factorial using loop

```python
n = int(input("Enter the number you want to find factorial of:"))
def factorial_loop(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial_loop(n))
```

### Factorial using recursion

```python
n1 = int(input("Enter the number you want to find factorial of:"))

def factorial_recursive(n1):
    if n1 < 0:
        return "Factorial is not defined for negative numbers"
    
    if n1 == 0:
        return 1
    else:
        return n1 * factorial_recursive(n1 - 1)


print(factorial_recursive(n1)) 
```
