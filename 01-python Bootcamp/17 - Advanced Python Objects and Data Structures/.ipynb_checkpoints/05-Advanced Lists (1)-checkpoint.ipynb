## extend

**append: appends whole object at end:**

```python
x = [1, 2, 3]
x.append([4, 5])
print(x)
```

**extend: extends list by appending elements from the iterable:**

```python
x = [1, 2, 3]
x.extend([4, 5])
print(x)
```
