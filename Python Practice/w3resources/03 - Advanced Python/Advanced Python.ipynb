###  1. Matrix Multiplication via List Comprehensions

Write a Python function that performs matrix multiplication using list comprehensions.

```python
def matrix_mul(A,B):
    if len(A[0]) != len(B):
        raise ValueError("Number of columns in A must equal number of rows in B")

    num_rows_A = len(A)
    num_cols_B = len(B[0])

    result = [[sum(A[i][k] * B[k][j] for k in range(len(B))) for j in range(num_cols_B)] for i in range(num_rows_A)]

    return result


if __name__ == "__main__":
    
    A = [[1, 2, 3],
         [4, 5, 6]]

    B = [[7, 8],
         [9, 10],
         [11, 12]]

  
    print(matrix_mul(A, B))
```

### 2. Class-Based Decorator for Execution Time
Write a Python program to create a class-based decorator that logs the execution time of methods.

```python
import time

class ExecutionTimer:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        start = time.time()
        result = self.func(*args, **kwargs)
        end = time.time()
        print(f"Execution time: {end - start:.6f} seconds")
        return result
```

```python
@ExecutionTimer
def sample_function():
    total = 0
    for i in range(1000000):
        total += i
    return total

sample_function()
```
