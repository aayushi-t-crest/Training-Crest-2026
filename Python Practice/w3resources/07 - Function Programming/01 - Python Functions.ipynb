##  Maximum of Three Numbers

Write a Python function to find the maximum of three numbers

```python
def max_three_nums(a,b,c):
    if ((a>b) and (a>c)):
        return a
    elif ((b>a) and (b>c)):
        return b
    else:
        return c

highest_num = max_three_nums(40,30,20)

print(f"Highest number is : {highest_num}")
```

## Check whether a number is perfect or not

```python
def check_num(n):
    sum = 0

    for x in range(1,n):
        if n % x == 0:
            sum += x

    return sum == n
```

```python
print(check_num(6))
```
